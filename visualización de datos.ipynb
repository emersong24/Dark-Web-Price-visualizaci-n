```python
# Importación librerías 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importación del DataFrame
filename = './Dark Web Price Index 2021.xlsx'
df = pd.read_excel(filename)
df.head(1)
```

```python
# Información general del DataFrame. No contiene datos nulos
df.info()
```

```python
# Eliminación de datos duplicados
df= df.drop_duplicates()
```

```python
# Descripción del DataFrame Filtrado
df.describe()
```

```python
# Producto más caro
df.max()
```

```python
# Producto más barato
df.min()
```

```python
# Grafíco de barras sobre categorías 
Categorias= df.groupby(['Category'])['Avg. dark web Price (USD)'].mean()
plt.style.use('seaborn')
plt.bar(Categorias.index, Categorias.values);
plt.ylabel('Promedio de precio en USD')
plt.title('Categorias más cotizadas en la Dark Web', fontsize = 16);
plt.xticks(rotation = 90);
```

```python
Productos= df.groupby(['Product'])['Avg. dark web Price (USD)'].sum().sort_values(ascending=False)
productos = Productos.head(10)
plt.barh(productos.index, productos.values)
plt.title('10 productos más costosos', fontsize = 16, horizontalalignment= 'right');
plt.ylabel('Producto')
plt.xlabel('Precio en USD', horizontalalignment= 'right');
```
